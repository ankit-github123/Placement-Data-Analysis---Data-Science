```python
placement_list = []
```

```python
for i in range(0, df.shape[0]):
  placement_list.append([str(df.values[i,j]) for j in range(0, df.shape[1])])
```

```python
placement_list
```

```python
from apyori import apriori
association_rules = apriori(transactions = placement_list, min_support = 0.005, min_confidence = 0.3, min_lift = 4, min_length = 3,max_length = 4)
```

```python
association_rules
```

```python
results=list(association_rules)
```
